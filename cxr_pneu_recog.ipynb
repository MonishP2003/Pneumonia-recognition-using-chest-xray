```python
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

Load the saved model

```python
model = tf.keras.models.load_model('cxr_pneu1.h5')
items = ['Normal', 'Bacterial pneumonia', 'Viral pneumonia']
```

Load the image

```python
img = cv2.imread(r"C:\Users\monis\Downloads\chest_xray\val\PNEUMONIA\person1950_bacteria_4881.jpeg",
                 0)
img = cv2.resize(img, (256, 256))
img = np.invert(np.array([img]))
prediction = model.predict(img)
print(np.argmax(prediction))
print("The condition is probably {}".format(items[np.argmax(prediction)]))
plt.imshow(img[0], cmap=plt.cm.binary)
plt.show()
```
